### Bool mask and bool indexing

```python
import pandas as pd
import numpy as np
```

```python
floats = pd.DataFrame(np.random.randn(7, 4), columns=["A", "B", "C", "D"])
floats
```

```python
# Creating a bool mask for column "B"
floats["B"] >= 0
# Selecting all rows where "B" >= 0
floats[floats["B"] >= 0]



# (Creating masks for rows is less common)

# Creating a bool mask for row 0 
floats.loc[0, :] >= 0

# Selecting all columns where row 0 >= 0
floats.loc[:, floats.loc[0, :] >= 0]
```

## Querying data
Filtering rows based on values of certain attributes can be done either with bool index or pandas query method.

```python
autos = pd.read_json("../Data/autos.json")

volvo_cars = autos[autos["make"] == "volvo"]
volvo_cars = autos.query("make == 'volvo'")

small_engines = autos[autos["engine-size"] <= 140]
small_engines = autos.query("`engine-size` <= 140")

small_volvos = autos[(autos["make"] == "volvo") & (autos["engine-size"] <= 140)]
small_volvos = autos.query("make == 'volvo' and `engine-size` <= 140")

small_engines_and_volvos = autos[(autos["make"] == "volvo") | (autos["engine-size"] <= 140)]
small_engines_and_volvos = autos.query("make == 'volvo' or `engine-size` <= 140")
```

```python
autos_copy = autos.copy()
my_mask = (autos_copy["make"] == "volvo") & (autos_copy["engine-size"] <= 140)
autos_copy.loc[my_mask, ["price"]] = 0
autos_copy.tail(10)
```

```python
import seaborn as sns

sns.relplot(x="engine-size", y="price", data=small_engines_and_volvos)
```

## Sorting DataFrame
DataFrames can be sorted using the `.sort_values()` method.

Parameters:
- `by` column, or list of columns
- `ascending` (optional), default is `True`

```python
autos.sort_values(by=["make", "engine-size"], ascending=[False, True])
```
