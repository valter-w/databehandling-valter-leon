### Convert to common formats

```python
# copy dataframe to clipboard (then use Ctrl+V)
autos.to_clipboard()
```

```python
autos.to_csv("../Data/autos.csv", index=None)
csv_str = autos.to_csv()
print(csv_str[:1000])
```

```python
autos.to_json("../Data/new_autos.json", orient="records")
json_str = autos.to_json(indent=4, orient="records")
print(json_str[:200])
```

```python
autos_dict = autos.to_dict(orient="records")
autos_dict[0]
```
